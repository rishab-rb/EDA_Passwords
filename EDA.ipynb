## PASSWORD STRENGTH

```python
strength_df = pd.read_csv('strength.csv', index_col=0)
strength_df
```

```python
labels = "WEEK","MODERATE","STRONG" 
colors = ['lightblue','lightsteelblue','silver']
explode = (0.01,0.01,0.1)

plt.pie(strength_df['frequency'], labels=labels, autopct='%1.1f%%', 
        startangle=15, shadow=True, colors=colors, explode=explode)
plt.title('PASSWORD STRENGTHS')
plt.axis('equal')
plt.show()
```

## LOWER LETTER FREQUENCY

```python
lower_letter_frequency_df = pd.read_csv('lower_letter_frequency.csv', index_col=0)
lower_letter_frequency_df
```
